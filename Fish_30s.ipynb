```python
import glob

label_files = glob.glob("/content/drive/MyDrive/Fish_30s/dataset/labels/train/*.txt")

for file in label_files:
    new_lines = []
    with open(file, "r") as f:
        for line in f:
            parts = line.strip().split()
            parts[0] = "0"   # force class = 0
            new_lines.append(" ".join(parts))
    with open(file, "w") as f:
        f.write("\n".join(new_lines))

print("✅ All labels fixed to class 0")
```

```python
import glob

label_files = glob.glob("/content/drive/MyDrive/Fish_30s/dataset/labels/**/*.txt", recursive=True)

for file in label_files:
    new_lines = []
    with open(file, "r") as f:
        for line in f.readlines():
            parts = line.strip().split()
            if len(parts) >= 5:
                parts[0] = "0"   # force fish class
                new_lines.append(" ".join(parts))
    with open(file, "w") as f:
        f.write("\n".join(new_lines))
```
